```python
import pandas as pd
import numpy as np
from pandas import DataFrame
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import preprocessing
import math
from sklearn.model_selection import train_test_split
from sklearn import metrics 
```

```python
data = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
data = data.drop(columns=['StandardHours','EmployeeCount','Over18','EmployeeNumber','StockOptionLevel'])

le = preprocessing.LabelEncoder()
categorial_variables = ['Attrition','BusinessTravel','Department','EducationField',
                        'Gender','JobRole','MaritalStatus','OverTime']
for i in categorial_variables:
    data[i] = le.fit_transform(data[i])
data.head(5)
data.to_csv('LabelEncoded_CleanData.csv')
```

```python
target = data['Attrition']
train = data.drop('Attrition',axis = 1)
train.shape
```

```python
#Defining a function which will give us train and test accuracy for each classifier.
def train_test_error(y_train,y_test):
    train_error = ((y_train==Y_train).sum())/len(y_train)*100
    test_error = ((y_test==Y_test).sum())/len(Y_test)*100
    train_accuracy.append(train_error)
    test_accuracy.append(test_error)
    print('{}'.format(train_error) + " is the train accuracy")
    print('{}'.format(test_error) + " is the test accuracy")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(train, target, test_size=0.33, random_state=42)
```

```python
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

class App(tk.Frame):
    def __init__(self, master=None):
        super().__init__(master)
        self.master = master
        self.master.title("Employee Attrition Prediction")
        self.pack()
        self.create_widgets()

    def create_widgets(self):
        self.select_file_button = tk.Button(self, text="Select Dataset", command=self.select_file)
        self.select_file_button.pack(pady=10)

        self.train_button = tk.Button(self, text="Train Model", command=self.train_model)
        self.train_button.pack(pady=10)

        self.predict_button = tk.Button(self, text="Predict Attrition", command=self.predict_attrition)
        self.predict_button.pack(pady=10)

    def select_file(self):
        file_path = filedialog.askopenfilename()
        if file_path.endswith(".csv"):
            self.df = pd.read_csv(file_path)
            messagebox.showinfo("Success", "Dataset Loaded Successfully")
        else:
            messagebox.showerror("Error", "Please select a CSV file")

    def train_model(self):
        if hasattr(self, "df"):
            le = LabelEncoder()
            self.df["Attrition"] = le.fit_transform(self.df["Attrition"])

            # One-hot encoding of categorical features
            cat_cols = self.df.select_dtypes(include=['object']).columns
            self.df = pd.get_dummies(self.df, columns=cat_cols)

            X = self.df.drop("Attrition", axis=1)
            y = self.df["Attrition"]

            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

            self.clf = LogisticRegression()
            self.clf.fit(X_train, y_train)

            messagebox.showinfo("Success", "Model Trained Successfully")
        else:
            messagebox.showerror("Error", "Please select a CSV file first")

    def predict_attrition(self):
        if hasattr(self, "clf"):
            # One-hot encoding of categorical features
            cat_cols = self.df.select_dtypes(include=['object']).columns
            df_predict = pd.get_dummies(self.df, columns=cat_cols)

            prediction = self.clf.predict(df_predict.drop("Attrition", axis=1))
            prediction = ["Yes" if p==1 else "No" for p in prediction]
            messagebox.showinfo("Prediction", f"Attrition Predictions:\n {prediction}")
        else:
            messagebox.showerror("Error", "Please train the model first")

root = tk.Tk()
app = App(master=root)
app.mainloop()
```
